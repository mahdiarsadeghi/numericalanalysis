# B-splines

```julia
using Plots
```

## $B(k,x)$

```julia
function B(k,x)
    if (x<k-1) return 0
        elseif (x>k+1) return 0 
        elseif (x<=k) return x-(k-1)
        elseif (x>k) return (k+1)-x
    end
end

x = -3:0.1:3
p = plot(title="Hat functions")
for k=-3:3
    plot!(p,x,B.(k, x), label=:false)
end
p
```

# Linear interpolation

```julia
function y(x,xk,yk)
    y = zeros(size(x))
    for k = 1:size(xk,1)
        y += yk[k]*B.(k-4,x)
    end
    return y
end

k = -3:3
x = sin.(k)
scatter(k,x,title="Interpolation using linear B-spline"
    , label="Knots",legend=:bottomright)
plot!(-3:0.1:3,y(-3:0.1:3,k,x), label="Linear B-spline")
plot!(-3:0.01:3, sin.(-3:0.01:3), label="Original Function")
```
