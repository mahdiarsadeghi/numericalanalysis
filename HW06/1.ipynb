# Power method for computing the eigenvector of matrix 

```julia
using Plots, LinearAlgebra
```

## 1

```julia
function nunitradius(N)
    u = 0:0.1:1
    p = plot(u, zeros(size(u,1)), label=:false, title="Unit vectors before iteration")
    for n=1:N-1
        ϕ = 2*π*n/N
        x = u.*cos(ϕ)
        y = u.*sin(ϕ)
        p = plot!(p,x,y,label=:false)
    end
    display(p)
end

N = 20
nunitradius(N)
```

## 2

```julia
function vectorinteration(A)
    N = 20

    u = 0:0.1:1
    v = zeros(2,N)
    n = 0:N-1
    ϕ = 2*π*n./N
    v[1,:] = cos.(ϕ)
    v[2,:] = sin.(ϕ)
    # iterative process
    tol = 1
    con = 0
    while tol>0.1
        nex = A*v
        tol = abs(norm(v-nex))
        v = nex
        for i=1:N
            v[:,i] = normalize(v[:,i])
        end
        con += 1
        if con>1e5
            break
        end
    end
    return v
end

A = [1 2; 2 4]
v = vectorinteration(A)

# show something
u = 0:0.1:1
p = plot(label=:false, title="Unit vectors before iteration")
for i=1:size(v,2)
    x = v[1,i]*u
    y = v[2,i]*u
    p = plot!(p,x,y,label=:false)
end
display(p)
```
