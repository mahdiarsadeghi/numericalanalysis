# PCA
This is a badass spider.

```julia
using CSV, Plots

data = Matrix(CSV.read("Datafile.csv"))
scatter(data[1,:],data[2,:], label=:false)
```

```julia
# upside down figure
scatter(data[1,:],data[2,:], data[3,:], label=:false)
```

## useful dimensions

Data has three usefull dimensions.

```julia
using MultivariateStats

x = fit(PCA, data)
```

```julia
projection(x)
```

```julia
p = projection(x)
t = transform(x,data)

scatter(t[1,:],t[2,:], label=:false)
```

```julia
# standing spider
scatter(t[1,:],t[2,:],t[3,:], label=:false)
```
