# SVD & Eigenvalue

```julia
using LinearAlgebra

function test(A)
    l = sqrt.(eigen(A'*A).values)
    r = reverse(svd(A).S) 
    if norm(l-r)<1e-10
        return 0
    end
    return 1
end
```

```julia
using Plots

n = 2:10
scatter(n, test.(rand.(n,n)), label="err", xlabel="dim(A)")
```
