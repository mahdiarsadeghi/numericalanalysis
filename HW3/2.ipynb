# Eigenvalues of $A$ based on $a$

```julia
using LinearAlgebra
function A(a)
    eigvals([1 a; a 1])
end

A(1)
```

$
\begin{align}
det(A-\lambda I) &= 0 \\
(1-\lambda)^2 - a^2 &= 0 \\
\lambda^2 - 2 \lambda +  (1- a^2) &=0 \\
\rightarrow \lambda &= \frac{1 \pm \sqrt{1 - 1 +a^2}}{1} \\
&= 1 \pm \| a\|
\end{align}
$

# Quadratic form

```julia
using Plots

function quadratic(u)
    range = collect(0:0.01:2)
    outpt = zeros(size(range,1))
    for n in 1:size(range,1)
        a = range[n]
        y =  u' * [1 a; a 1] * u
        outpt[n] = y
    end
    plot(range, outpt, xlabel="a", ylabel="f(u)"
        , title="u="*string(u), label=:false, linewidth=4)
end

quadratic([5, 2])
```

For $a=1$, the output of $f(u)$ function is $(u[1]+u[2])^2$;
