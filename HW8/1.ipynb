## $L$ and $\theta$

```julia
function ladder(θ)
    a, b, α = [4,3,π/12]
    h₁ = b/sin(θ)
    h₂ = a/sin(-θ-α+π/2)
    return h₁+h₂
end

θ = 0.1:1e-2:1.2
plot(θ, ladder.(θ), xlabel="θ(rad)", linewidth=4, 
    ylabel="Length of ladder", label="Numerical Solution")
```

```julia
minimum(ladder.(θ))
```

# $\alpha = 0$ 

In the case where $a=b=3$, we expect the minimum $l(\theta)$ to be $\frac{6}{\sin(\pi/4)}$.

```julia
2*3/sin(π/4)
```

```julia
function test(θ)
    a, b, α = [3,3,0]
    h₁ = b/sin(θ)
    h₂ = a/sin(-θ-α+π/2)
    return h₁+h₂
end

θ = 0.18:1e-2:1.4
plot(θ, test.(θ), xlabel="θ(rad)", linewidth=4, 
    ylabel="Length of ladder", label="Numerical Solution")
```

```julia
minimum(test.(θ))
```
