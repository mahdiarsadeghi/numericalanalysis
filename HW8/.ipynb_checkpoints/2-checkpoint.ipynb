# Root Finder

```julia
using Plots

f(x) = x^4+0.4*x^3-0.5*x^2+1.2*x-1.5 
range = -100:1:100
plot(range, f.(range), linewidth=4, label="p(x)")
```

## Vieta's formulas for $n=4$:

$
\begin{align}
x_1 + x_2 + x_3 + x_4 &= -0.4 \\
x_1 (x_2 + x_3 + x_4) + x_2 (x_3 + x_4) + x_3 x_4 &= -0.5 \\
x_1 x_2 ( x_3 + x_4) + x_2 x_3 x_4 + x_1 x_3 x_4   &= -1.2 \\
x_1 x_2 x_3 x_4 &= 1.5
\end{align}
$

## Jacobian Matrix

$
\begin{bmatrix}
  1 & 1 & 1 & 1 \\
  x_2 + x_3 + x_4 & x_1 + x_3 + x_4 & x_1 + x_2 + x_4 & x_1 + x_2 + x_3 \\
  x_2x_3 + x_2x_4 + x_3x_4 & x_1x_3 + x_1x_4 + x_3x_4 & x_1x_2 + x_2x_4 + x_1x_4 & x_1x_2 + x_2x_3 + x_1x_3 \\
  x_2x_3x_4 & x_1x_3x_4 & x_1x_2x_4 & x_1x_2x_3
\end{bmatrix}
$

# Solution with Roots

```julia
using Roots

p = find_zeros(f, -100, 100)
```

```julia
# test
f.(p)
```

# Solution with PolynomicalRoots

```julia
using PolynomialRoots

p = roots([-1.5, 1.2, -0.5, 0.4, 1])
```

```julia
# test
f.(p)
```
