# Read WAV file

```julia
using WAV
sound, freq = wavread("OdeToJoy.wav")
freq = Integer(freq)
```

# Visualization

```julia
using Plots, FFTW
# create time axis
time = (0:(size(sound,1)-1)) / freq
# Fourier Transform of it 
F = fft(sound) |> fftshift
freqs = fftfreq(length(time), freq) |> fftshift
# plots 
time_domain = plot(time, sound, title = "Signal", xlim=(1, 8.5), label=:false)
freq_domain = plot(freqs, abs.(F), title = "Spectrum", xlim=(0, 2000), label=:false) 
plot(time_domain, freq_domain, layout = (1,2))
```

First peak from the left hand side is located at f=291.4 Hz. Therefore, <b>D</b> is the music key for this tune.
