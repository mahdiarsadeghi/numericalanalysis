## 1

```julia
function ln2(n)
    x = ones(n)/2
    y = 1/2
    for k in 1:n-1
        x[k+1] = x[k]/2
        y += x[k+1]/(k+1)
    end
    y
end 

ln2(100)
```

## 2

```julia
using Plots 
N = exp2.(collect(1:2:200))
x = zeros(size(N))
for n in 1:size(N,1)
    x[n] = log(2) - ln2(n)
end 
display(log(2))
plot(N, x, xaxis=:log, yaxis=:log, seriestype=:scatter)
```
