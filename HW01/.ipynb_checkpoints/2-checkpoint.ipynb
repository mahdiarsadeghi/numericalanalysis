## mylog()

```julia
function mylog(z)
    x = z-1
    s = 1
    y = 0
    k = 1
    while (abs(x/k)>=1e-15 && k<1e7)
        y += s*x/k
        k += 1
        x *= z-1
        s *= -1
    end
    y
end 
```

## test

```julia
using Plots
z = collect(0.01:0.1:1.99)
e = z-exp.(mylog.(z))
plot(z, e, title="Round-Trip Test", ylabel="z-exp(y)",
    seriestype=:scatter, label=:none, xlabel="z")
```
