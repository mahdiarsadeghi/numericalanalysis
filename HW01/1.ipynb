## 1

```julia
function ln2(n)
    x = 2*ones(n)
    y = 0.5
    for k in 1:n-1
        x[k+1] = x[k]*2
        y += 1/(x[k+1]*(k+1))
    end
    y
end 

ln2(10)
```

## 2

```julia
using Plots 
N = 100
x = zeros(N)
for n in 1:N
    x[n] = log(2) - ln2(n)
end 
display(log(2))
plot(1:N, x, xaxis=:log, yaxis=:log, seriestype=:scatter,
        title="Error in log(2) series computation",
        label=:none)
```
