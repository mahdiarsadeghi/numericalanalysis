## Wallis method

```julia
function myln(z)
    # initialize computation
    b0, b, a  = [0, 1, z-1]
    Ak2, Ak1, Bk2, Bk1 = [1, b0, 0, 1]
    A, B = [b*Ak1+a*Ak2, b*Bk1+a*Bk2]
    # error tolerance
    ynm1, tol = [0, 1e-10]
    for k in 0:1e5
        yn = A/B
        diff = yn-ynm1
        if (abs(diff)<tol)
            return yn
        end
        ynm1 = yn
        Ak2, Ak1, Bk2, Bk1 = [Ak1, A, Bk1, B]
        ak = a*(kÃ·2)^2
        bk = b+1
        A, B = [b*Ak1+ak*Ak2, b*Bk1+ak*Bk2]
    end
    println("myln() failed to converge")
end 
```

## test

```julia
using Plots
z = collect(0.01:0.5:15)
e = z-exp.(myln.(z))
plot(z, e, title="Round-Trip Test", ylabel="z-exp(y)",
    seriestype=:scatter, label=:none, xlabel="z")
```
