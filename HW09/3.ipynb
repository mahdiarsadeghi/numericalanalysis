# $sin(2\pi x)e^x$ Interpolation 

```julia
f(x) = sin(2*π*x)*exp(x)

x = 0:.01:5
p = plot(x, f.(x), label="Continuous Function", linewidth=3
    , legend=:bottomleft)
```

#  Chebyshev Interpolation

## Sampling

```julia
N = 19
i = 0:N
c = cos.((2i.+1)*π/(2*N+2))
x = 2.5.+2.5*c
p = scatter!(p, x,f.(x), label="Sample Points"
    , title="Chebyshev Interpolation")
```

## Interpolation

```julia
function langrange(x,f,n)
    a = range(x[1],stop=x[end],length=n)
    b = zeros(size(a))
    for j = 1:size(x,1)
        l = ones(size(a))
        for m = 1:size(x,1)
            if (j!=m)
                l = l.*(a.-x[m])/(x[j]-x[m])
            end
        end
        b = b + f[j]*l
    end
    return [a,b]
end

a, b = langrange(x,f.(x),100)
plot!(p, a, b, label="Equispace Interpolation"
    , linestyle=:dash, linewidth=3)
```

#  Equispace Interpolation

## Sampling

```julia
x = 0:.01:5
p = plot(x, f.(x), label="Continuous Function", linewidth=3
    , legend=:bottomleft, title="Equispace Interpolation")

N = 19
x = range(0,stop=5,length=N)
scatter!(p, x,f.(x), label="Equispace Sample Points")
```

## Interpolation

```julia
function langrange(x,f,n)
    a = range(x[1],stop=x[end],length=n)
    b = zeros(size(a))
    for j = 1:size(x,1)
        l = ones(size(a))
        for m = 1:size(x,1)
            if (j!=m)
                l = l.*(a.-x[m])/(x[j]-x[m])
            end
        end
        b = b + f[j]*l
    end
    return [a,b]
end

a, b = langrange(x,f.(x),100)
plot!(p, a, b, label="Equispace Interpolation"
    , linestyle=:dash, linewidth=3)
```
