# $U x = b$ Solver

## Backsubsitution

```julia
function backsubstitution(u,b)
    x = zeros(size(b))
    for j in 1:size(b,1)
        j = size(b,1)+1-j
        x[j]  = b[j]
        for i in j+1:size(b,1)
            x[j] -= u[j,i]*x[i]
        end
        x[j] /= u[j,j]
    end
    return x
end
```

## Test

```julia
using LinearAlgebra, Plots

function test(n)
    L,U,p = lu(rand(n,n))
    b = rand(n)
    r = norm(b-U*backsubstitution(U,b))
    if r<1e-10 r=0 end
end

scatter(1:10,test.(1:10), label=:false, xlabel="n", ylabel="norm(r)")
```
