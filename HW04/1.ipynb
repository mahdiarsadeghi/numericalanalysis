# Load and visualize csv files

Note, images will show up in a new tab.

```julia
using CSV, ImageView
husky = Matrix(CSV.read("side_husky.csv"));
board = Matrix(CSV.read("checkerboard.csv"));
imshow(husky)   
imshow(board)
```

# SVD

```julia
using LinearAlgebra, Plots
x = svd(husky)
y = svd(board)
p1 = plot(x.S, yaxis=:log, xlabel="Index of singular value", label=:false,
          ylabel="Singular value", title="Side Husky", linewidth=3)
p2 = plot(y.S, yaxis=:log, xlabel="Index of singular value", label=:false,
          ylabel="Singular value", title="Checkboard", linewidth=3)
plot(p1,p2, layout=(1,2))
```
