# Chemometrics

```julia
using CSV, Plots
```

## Reference

```julia
rf = CSV.read("ref_spectra.csv")

plot(rf[:,1], rf[:,2], title="Reference spectra")
plot!(rf[:,1],rf[:,3])
plot!(rf[:,1],rf[:,4])
```

## Measured

```julia
df = CSV.read("unknown_mixture_spectrum.csv")

plot(df[:,1],df[:,2], label=false, 
    title  = "Measured optical absorption",
    xlabel = "Wavelength (ν)",
    ylabel = "Absoption")
```

# Linear Regression

```julia
# double check frequencies
df[:,1]==rf[:,1]
```

```julia
A = Matrix(rf[:,2:4])
s = df[:,1]
(A'*s)\(A'*A)
```
