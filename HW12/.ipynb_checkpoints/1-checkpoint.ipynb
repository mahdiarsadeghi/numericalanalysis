```julia
z = 0.1:0.1:10
plot(z, gamma.(z), label="Julia's Built in Gamma", 
    xlabel="z", ylabel="Γ(z)", linewidth=4)
plot!(z, gausslaguerre.(z), label="Gauss-Laguerre quadrature",
    linestyle=:dash, linewidth=3)
```

```julia
z = 0.1:0.01:1.0
scatter(z, gamma.(z), label="Julia's Built in Gamma", 
    xlabel="z", ylabel="Γ(z)")
scatter!(z, gausslaguerre.(z), label="Gauss-Laguerre quadrature")
```
